```python
seoul_complain.head()
```

```python
top_20_dong = seoul_complain['Dong'].value_counts().head(20)

fig = go.Figure(data=[go.Pie(labels=top_20_dong.index, values=top_20_dong.values)])


fig.show()
```

```python
city_counts = seoul_complain['City'].value_counts()
fig = go.Figure(data=[go.Bar(x=city_counts.index, y=city_counts.values,
                              marker=dict(color=px.colors.sequential.Inferno))])

fig.update_layout(
    title='서울 구별 민원 발생 횟수',
    xaxis=dict(title='서울 구별'),
    yaxis=dict(title='민원 발생 횟수')
)

fig.show()
```

```python
Dong_counts = seoul_complain['Dong'].value_counts()

fig2 = go.Figure(data=[go.Bar(x=Dong_counts.index, y=Dong_counts.values,
                             marker=dict(color=px.colors.sequential.Viridis))])

fig2.update_layout(
    title='서울 동별 민원 발생 횟수',
    xaxis=dict(title='서울 동별'),
    yaxis=dict(title='민원 발생 횟수')
)

fig2.show()
```

```python
import plotly.graph_objs as go
import plotly.express as px

Dong_counts = seoul_complain['Dong'].value_counts().reset_index()
Dong_counts.columns = ['Dong', 'Count']

fig2 = go.Figure(data=[go.Bar(y=Dong_counts['Dong'], x=Dong_counts['Count'], orientation='h',
                             marker=dict(color=px.colors.sequential.Viridis))])

fig2.update_layout(
    title='서울 동별 민원 발생 횟수',
    xaxis=dict(title='민원 발생 횟수'),
    yaxis=dict(title='서울 동별')
)

fig2.show()
```
