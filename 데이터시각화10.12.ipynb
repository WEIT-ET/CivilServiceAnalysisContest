```python
def address_to_dataframe(df):
    # 주소 열을 선택
    addresses = df['ADDRESS']

    # 새로운 열을 저장할 리스트 초기화
    provinces = []
    cities = []
    dong = []

    for address in addresses:
        address = str(address)
        parts = address.split()

        # 주소 문자열이 충분한 구성 요소를 가지고 있는지 확인
        if len(parts) >= 3:
            provinces.append(parts[0])
            if parts[1].endswith('시'):
                cities.append(parts[2])
            else:
                cities.append(parts[1])
            if len(parts) >= 4:
                dong.append(parts[3])
            else:
                dong.append(parts[2])

        else:
            # 주소 문자열이 3개의 요소 미만인 경우에는 빈 값으로 처리
            provinces.append('')
            cities.append('')
            dong.append('')

    # 새로운 열을 데이터프레임에 추가
    df['Province'] = provinces
    df['City'] = cities
    df['Dong'] = dong

    # provinces와 cities를 결합하여 새로운 파생 변수 생성
    df['Location'] = df['Province'] + ' ' + df['City']

    df = df.drop('ADDRESS', axis=1)
    return df
```

## 음주운전 데이터 추출

```python
Alcohol_accident=pd.read_csv('/content/drive/MyDrive/weit/음주운전사고.csv',encoding='cp949')
```

```python
Alcohol_accident = Alcohol_accident.drop(['다발지역폴리곤','시도시군구명','지점코드','사고다발지fid'], axis=1)
```

```python
def Alcohol_change(df):
  seoul_df = df['지점명'].str.contains('서울특별시', na=False)
  seoul_df = df[seoul_df]
  addresses = seoul_df['지점명']
  provinces = []
  cities = []
  dong = []

  for address in addresses:
        parts = address.split('(')
        address_without_brackets = parts[0]
        parts = address_without_brackets.split()
        provinces.append(parts[0])
        cities.append(parts[1])
        dong.append(parts[2][:3])

  seoul_df['Province'] = provinces
  seoul_df['City'] = cities
  seoul_df['Dong'] = dong
  seoul_df = seoul_df.drop('지점명', axis=1)
  return seoul_df
```

```python
Alcohol_accident= Alcohol_change(Alcohol_accident)
```

```python
Alcohol_accident.to_csv('서울교통사고데이터',index=False,encoding="utf-8-sig")
```
